```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl


df = pd.read_csv('s4 v1 particle 1 200 slices.csv')
data = df.to_numpy()

r_sq_s4_v1_p1=[]
t=[]

df.head()
x= df.iloc[:,5]
y= df.iloc[:,6]
t=df.iloc[:,7]
# print(y)

x_0= x[0]
y_0= y[0]

for i in range (len(x)):
    r_sq_s4_v1_p1.append((x[i] - x_0)**2 + (y[i] - y_0)**2)


print(len(r_sq_s4_v1_p1))
# print(r_sq)

plt.scatter(t, r_sq_s4_v1_p1)
```

```python
t= np.linspace(0,199,199)

# scatter = plt.scatter(x, y, c=t, cmap='viridis', s=20)

# # Add a colorbar for time
# cbar = plt.colorbar(scatter)
# cbar.set_label('Time')

# plt.xlabel('x')
# plt.ylabel('y')
# plt.title('Trajectory with Time Color Map')
# plt.show()


from matplotlib.collections import LineCollection

t = np.linspace(0, 199, 199)

# Prepare segments for a colored line
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

fig, ax = plt.subplots(figsize=(6,5))

lc = LineCollection(segments, cmap='inferno', norm=plt.Normalize(t.min(), t.max()))
lc.set_array(t)        # Assign time as color variable
lc.set_linewidth(2)

line = ax.add_collection(lc)
cbar = fig.colorbar(line, ax=ax)
cbar.set_label('Time')

ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
ax.set_xlabel('x($\mu m$)')
ax.set_ylabel('y ($\mu m$)')
# ax.set_title('Trajectory with Time Color Gradient')
plt.savefig('Trajectory with Time Color Gradient')
plt.show()
```

### Finding avg r^2 for each time step

```python
r_sq_master_list= [r_sq_s4_v1_p1, r_sq_s4_v1_p2, r_sq_s4_v2_p1, r_sq_s4_v2_p2,r_sq_s4_v3_p1,r_sq_s4_v4_p1,r_sq_s4_v4_p2,r_sq_s5_v1_p2, r_sq_s7_v1_p2, r_sq_s7_v2_p1,r_sq_s7_v2_p2,r_sq_s9_v1_p1,r_sq_s9_v1_p2,r_sq_s9_v2_p1]
print(len(r_sq_master_list))
#without s5 v1 p1, s7 v1 p1, s7 v1 p3, s7 v1 p4


min_len = min(len(d) for d in r_sq_master_list)
print(min_len)
truncated_r_sq = np.array([d[:min_len] for d in r_sq_master_list])

avg_r_sq = np.mean(truncated_r_sq, axis=0)
std_r_sq = np.std(truncated_r_sq, axis=0)

avg_std_r_sq=np.mean(std_r_sq)

print(avg_std_r_sq)

print(len(avg_r_sq))
t_final= np.arange(1, 199)

fit=np.polyfit(t_final, avg_r_sq,1)
bestfitline=np.polyval(fit, t_final)

slope=fit[0]
intercept=fit[1]

print(f"Best fit line equation: y = {slope:.2f}x + {intercept:.2f}")

plt.scatter(t_final, avg_r_sq, label='data')

plt.errorbar(t_final, avg_r_sq, yerr=std_r_sq, label='error bars')

plt.xlabel('time (s)')
plt.ylabel('average r square ($\mu m$)^2')
plt.plot(t_final, bestfitline,color='red', label='trendline')
plt.legend(loc='upper left')
plt.savefig('Particle Ensemble')
plt.show()
```
