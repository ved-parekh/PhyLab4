### Calculating Uncertainty in Slope of MnSO4

```python
import numpy as np

# Given data
dh_avg_mnso4 = [
    0,
    0.00009,
    0.00029,
    0.0006,
    0.00118,
    0.00181,
    0.00253,
    0.00336,
    0.00431,
    0.00526,
    0.00625,
    0.00709,
    0.00793
]

std_mnso4 = [
    0,
    0.000005773502692,
    0.00002081665999,
    0.0001006644591,
    0.00009291573243,
    0.00006506407099,
    0.0001101514109,
    0.00007371114796,
    0.00008,
    0.000144222051,
    0.00004932882862,
    0.00009,
    0.00006027713773
]

mag_field = [
    0.01282, 0.0749275, 0.137035, 0.1991425, 0.26125,
    0.3233575, 0.385465, 0.4475725, 0.50968,
    0.5717875, 0.633895, 0.6960025, 0.75811
]

mag_field_squared = [x**2 for x in mag_field]

x = np.array(mag_field_squared)
y = np.array(dh_avg_mnso4)
sigma = np.array(std_mnso4)

# Prevent infinite weight for sigma=0
sigma[sigma == 0] = min(s for s in sigma if s>0)

w = 1/sigma**2

# Weighted least squares formulas
W = np.sum(w)
Wx = np.sum(w*x)
Wy = np.sum(w*y)
Wxx = np.sum(w*x*x)
Wxy = np.sum(w*x*y)

Delta = W*Wxx - Wx**2

slope = (W*Wxy - Wx*Wy) / Delta
intercept = (Wxx*Wy - Wx*Wxy) / Delta

# Uncertainties
slope_error = np.sqrt(W / Delta)
intercept_error = np.sqrt(Wxx / Delta)

print("Slope =", slope)
print("Slope uncertainty =", slope_error)
print("Intercept =", intercept)
print("Intercept uncertainty =", intercept_error)
```
