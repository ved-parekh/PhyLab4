```python
from scipy.stats import norm


x_master_list= [x_s4_v1_p1,x_s4_v1_p2,x_s4_v2_p1,x_s4_v2_p2,x_s4_v3_p1,x_s4_v4_p1,x_s4_v4_p2,x_s5_v1_p1,x_s5_v1_p2,x_s7_v1_p1,x_s7_v1_p2,x_s7_v1_p3,x_s7_v1_p4,x_s7_v2_p1,x_s7_v2_p2,x_s9_v1_p1,x_s9_v1_p2,x_s9_v2_p1]

# Combine all data into one array
x_all = np.concatenate(x_master_list)
print(len(x_all))
# Compute statistics
mean = np.mean(x_all)
std_dev = np.std(x_all)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot normalized histogram
count, bins,_ = plt.hist(x_all, bins=10, density=True, alpha=0.8, label="Data Histogram")

# Generate points for the Gaussian curve
x = np.linspace(min(x_all), max(x_all), 300)
gaussian = norm.pdf(x, mean, std_dev)

# Plot the Gaussian fit
plt.plot(x, gaussian, 'r-', linewidth=2, label="Gaussian Fit")

# Labels
plt.xlabel("Step Length ($\mu m$)")
plt.ylabel("Probability Density")
plt.title("Normalized Histogram with Gaussian Fit")
plt.legend()
plt.savefig('Histogram of Step Length')
plt.show()





```
