### 0.6 kV



```python
import numpy as np
import matplotlib.pyplot as plt
import math

def radius(terminal_v):
    v = terminal_v * 1e-3
    delta_rho = 893.8      # kg/m^3 (This is ρ - ρ_air)
    p = 1.01e5             # Pa
    eta = 1.844e-5         # Pa·s
    b = 8.20e-3            # m·Pa
    g = 9.81               # m/s^2

    # Formula components
    term1_sqrt = (b**2) / (4 * p**2)
    term2_sqrt = (9 * eta * v) / (2 * g * delta_rho)
    term_subtracted = b / (2 * p) # <-- This was the corrected part

    # Final radius calculation
    r = np.sqrt(term1_sqrt + term2_sqrt) - term_subtracted
    
    print(r) # You can uncomment this to see the output
    

def charge(v_prime, v, r, voltage):
    v_prime =v_prime*1e-3 # example value in m/s
    v = v*1e-3        # example value in m/s
    r = r # radius in meters
    voltage=voltage
    rho = 893.8        # density of oil in kg/m^3
    rho_air = 1.225    # density of air in kg/m^3
    g = 9.81           # acceleration due to gravity in m/s^2
    d=0.01
    # Calculate q
    term1 = (v_prime / v) + 1
    term2 = (4 * math.pi * r**3 * d * (rho - rho_air) * g) / (3 * voltage)
    q = term1 * term2
    print("Charge q =", q, "Coulombs")

radius(0.17)
charge(0.17, 0.07, 1.22e-6, 600)
```
